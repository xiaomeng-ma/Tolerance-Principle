```python
import openturns as ot
data = pd.read_csv('8x4.csv')
y = data['Adam'].dropna()
x = np.array(range(1, len(y)+1))
```

```python
def Harmonic(N,s,q):
    total = 0.0
    for i in range(1,N+1):
        total += 1.0/((i+q)**s)
    return total
```
